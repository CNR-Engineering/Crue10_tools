<h3> LdE 16_11_2002 </h3>

```python
Cc = 'Cc_P020_16_11_2002'
min_section = 3
branches,info_res=Info_calcul(results,Cc,modele,min_section)
```

```python
sections='nbS'
indices= 'ind'
hf_df = pd.DataFrame({elt: branches[elt].hf for elt in branches})
Br_df = pd.DataFrame({elt: branches[elt].Br for elt in branches})
hmoy_df = pd.DataFrame({elt: branches[elt].hmoy for elt in branches})
hmoy_fp_df = pd.DataFrame({elt: branches[elt].hmoy_FP for elt in branches})

nb_sections_total = [info_res[elt+sections] for elt in branches]
successives= [info_res[elt+indices]==list(branches[elt].index) for elt in branches]
colors= []

for elt in successives:
    if elt:
        colors.append('royalblue')
    else:
        colors.append('darkorange')
```

```python

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10))

# hmoy
bp = hmoy_df.boxplot(grid=False, medianprops=dict(linewidth=1.8),rot=45,ax=axes[0,0])
bp.set_ylabel(r'$h_{moy}$ (m)', fontsize=20)
x = np.arange(hmoy_df.shape[1])

noofobs = hmoy_df.notnull().sum().astype(int)

for tick,label in zip(x,bp.get_xticklabels()):
    bp.text(x[tick]+1, hmoy_df.median()[tick] + .1,str(noofobs[tick])+'/'+str(nb_sections_total[tick]),horizontalalignment='center', color=colors[tick], weight='bold', fontsize=13)
bp.tick_params(axis='both',labelbottom=False)

# hfmoy
bp2 = hmoy_fp_df.boxplot(grid=False, medianprops=dict(linewidth=1.8),rot=45,ax=axes[1,0])
bp2.set_ylabel(r'$h_{fmoy}$ (m)', fontsize=20)
x = np.arange(hmoy_fp_df.shape[1])

noofobs = hmoy_fp_df.notnull().sum().astype(int)

for tick,label in zip(x,bp2.get_xticklabels()):
    bp2.text(x[tick]+1, hmoy_fp_df.median()[tick] + .1, str(noofobs[tick])+'/'+str(nb_sections_total[tick]),horizontalalignment='center', color=colors[tick], weight='bold', fontsize=13)
    
    
# Br
bp3 = Br_df.boxplot(grid=False, medianprops=dict(linewidth=1.8),rot=45,ax=axes[0,1])
bp3.set_ylabel(r'$B_r$', fontsize=20)
x = np.arange(Br_df.shape[1])

noofobs = Br_df.notnull().sum().astype(int)

for tick,label in zip(x,bp3.get_xticklabels()):
    bp3.text(x[tick]+1, Br_df.median()[tick] + .05, str(noofobs[tick])+'/'+str(nb_sections_total[tick]),horizontalalignment='center', color=colors[tick], weight='bold', fontsize=13)
bp3.tick_params(axis='both',labelbottom=False) 

# hf
bp = hf_df.boxplot(grid=False, medianprops=dict(linewidth=1.8),rot=45, ax=axes[1,1])
bp.set_ylabel(r'$h_f$ (m)', fontsize=20)
x = np.arange(hf_df.shape[1])

noofobs = hf_df.notnull().sum().astype(int)

for tick,label in zip(x,bp.get_xticklabels()):
    bp.text(x[tick]+1, hf_df.median()[tick] + .1, str(noofobs[tick])+'/'+str(nb_sections_total[tick]),horizontalalignment='center', color=colors[tick], weight='bold', fontsize=13)
    
```
